<b>Problem Statement: </b><br><br>
   Given data for thousands of restaurants in Bangalore,find insights and suggestions to improve the overall restaurant experience.
   
<b> Insights and suggestion Summary</b><br><br>
The following insights/suggestions are found on the basis of various charts and table manipulation which can be found in subsequent blocks of code:
   1. Koramangala(approx 300) and BTM(approx 225) area has the most restaurants in the Silicon Valley of India. 
   2. The average footfall of top 20 locations is 321.

   4. It's better to put your food place on Zomato/Swiggy and gather more customers as restaurants with online orders facility account for 60% traffic.

   7. Favorable locations to open new restaurants should be Church Street, Lavelle Road, Indiranagar, St. Marks Road as they don't have enough restaurants and the competition is also really less although a lot of potential customers.
   8. The Top-4 favorite cuisines for Bangloreans are North Indian, Chinese, Continental and Italian. Although more Italian restaurants should be established.
   9. Full meal restaurants should have salad as an option as people tend to like them but not many restaurants are serving them.

```python
data = pd.read_csv('RestoInfo.csv')
```

<b>Analysis of Restaurants in Bangalore</b>

<br><br>
But are all these clusterings valid? It maybe possible that some locations have very less restaurants but more footfall. Let's find out. Average footfall of restaurants in locations can be used as metric. <br><br> I am taking a general assumption that the number of votes are directly proportional to the footfall.

```python
x = pd.DataFrame(data["location"].value_counts()).reset_index().rename(columns = {"location":"No. of Restaurant", "index": "location"})
y = pd.DataFrame(data.groupby("location").aggregate({"votes":"sum"})).reset_index().rename(columns = {"index": "location"})
y = pd.merge(y, x, on = "location")
y["Average Footfall per restaurant"] = y["votes"]/y["No. of Restaurant"]
y = y.sort_values(by = "Average Footfall per restaurant", ascending = False)
```

```python
top_20 = list(data.location.value_counts().iloc[:20].index)
avg_top_20 = np.average(y[y["location"].isin(top_20)]["Average Footfall per restaurant"])
print(f"The average footfall of top 20 locations based on the number of restaurants is {avg_top_20}")
```

```python
y.head(10)
```

Hence, it can be evidently said that favorable locations to open new restaurants should be Church Street, Lavelle Road, Indiranagar, St. Marks Road as they don't have enough restaurants and the competition is also really less although a lot of potential customers.

Now let's look at Cuisines and Ratings analysis

```python
uniq_cuisines = set()
data["number of cuisines"] = 0
for i in range(data.shape[0]):
    uniq_cuisines = uniq_cuisines.union(set(data["cuisines"][i].split(",")))
    data["number of cuisines"][i] = len(data["cuisines"][i].split(","))
```

```python
print(f"Total number of unique cuisines that are being served in Bangalore are {len(uniq_cuisines)}")
```

```python
x = pd.DataFrame(data["number of cuisines"].value_counts()).reset_index().rename(columns = {"index": "no of cuisines", "number of cuisines": "no of rest"})
x["Perc"] = (x["no of rest"]/sum(x["no of rest"]))*100
labels = x["no of cuisines"]
size = x["Perc"]
plt.figure(figsize = (10,6))
plt.title("Market Size of Restaurants by number of cuisines")
g = plt.pie(size, labels = labels, autopct='%1.1f%%');
```

More than 80% of restaurants have 3 or less cuisines!!

```python
x = pd.DataFrame(data.groupby('online_order').aggregate({"votes": "mean"})).reset_index().rename(columns = {"votes": "Average footfall"})
plt.figure(figsize = (10, 6))
labels = list(x["online_order"])
size = list(x["Average footfall"])
plt.title("Average Footfall size for Online Order")
plt.pie(size, labels = labels, autopct='%1.1f%%', explode = [0,0.1]);
```

This is proved from the data and was expected too. People like to order more than to dine out. Hence it's better to put your food place on Zomato/Swiggy and gather more customers. 

```python
x = data[data["rate"].notnull()]
x = x[x["rate"]!="NEW"]
x = x[x["rate"]!="-"]
x["rate"] = x["rate"].apply(lambda x: float(x[:-2].strip()))
```

```python
plt.figure(figsize = (10,6))
plt.title("Table booking Options vs. Ratings")
sns.violinplot(x = 'book_table', y = 'rate', data = x)
```

It can be evidently seen from the above plot that when the restaurants have an option to book table, people like it more, so it is advisable for food places to give this option. It is interesting to see why table booking apps are taking over the market swiftly.

```python
y = x.groupby('number of cuisines').aggregate({"rate": "mean"}).reset_index()
z = x["number of cuisines"].value_counts().reset_index().rename(columns = {"number of cuisines": "Count", "index": "number of cuisines"})
y = pd.merge(y, z, on = "number of cuisines")
fig, ax = plt.subplots(1, 2, figsize = (10, 6))
#plt.figure(figsize = (10,6))
ax[0].set_title("Number of cuisines vs Average Ratings")
ax[1].set_title("Number of cuisines vs Number of such restaurants")
sns.regplot(ax = ax[0], x = "number of cuisines", y = "rate", data = y)
sns.regplot(ax = ax[1], x = "number of cuisines", y = "Count", data = y)
```

A new inference can be derived that multi-cuisine restaurants are liked more but, the number of such restaurants are lowers. The reason might be that multi-cuisine restaurants are costly to build and maintain. So the safe option is to open 4-cuisine restaurants which might not be that costly and also liked by customers.

```python
cuisines = defaultdict()
ratings = defaultdict()
votes = defaultdict()
x = x.reset_index(drop = True)
for i in range(x.shape[0]):
    cuis = list(x["cuisines"][i].split(", "))
    for cui in cuis:
        if cui not in cuisines.keys():
            cuisines[cui] = 1
            ratings[cui] = x["rate"][i]
            votes[cui] = x["votes"][i]
        else:
            cuisines[cui] +=1
            ratings[cui] += x["rate"][i]
            votes[cui] += x["votes"][i]

cuisines = pd.DataFrame.from_dict({"Cuisines": list(cuisines.keys()), "No of Restaurants": list(cuisines.values()), "Average Rating":list(ratings.values()), "Footfall": list(votes.values())})
cuisines["Average Rating"] = cuisines["Average Rating"]/cuisines["No of Restaurants"]
```

```python
cuisines = cuisines.sort_values(by = "Footfall", ascending = False)
cuisines.head(20)
```

The Top-4 favorite cuisines for Bangloreans are North Indian, Chinese, Continental and Italian. Although more Italian restaurants should be established.<br>
Also full meal restaurants should have salad as an option as people tend to like them but not many restaurants are serving them.
